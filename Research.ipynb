```
import numpy as np

# compute sigmoid nonlinearity
def sigmoid(x):
    output = 1/(1+np.exp(-x))
    return output

# convert output of sigmoid function to its derivative
def sigmoid_output_to_derivative(output):
    return output*(1-output)
    
# input dataset
X = np.array([  [0,1],
                [0,1],
                [1,0],
                [1,0] ])

print("X:",X)
    
# output dataset            
y = np.array([[0,0,1,1]]).T
print("y",y)
print("y not T",np.array([[0,0,1,1]]))

# seed random numbers to make calculation
# deterministic (just a good practice)
np.random.seed(1)

# initialize weights randomly with mean 0
synapse_0 = 2*np.random.random((2,1)) - 1
print("Random: {}",synapse_0)

# layer_0 = X
# layer_1 = sigmoid(np.dot(layer_0,synapse_0))
# print("Sigmod: {}",layer_1)

# layer_1_error = layer_1 - y
# print("Error: {}",layer_1_error)


for iter in range(10):
    print("________________________________") 
    # forward propagation
    layer_0 = X
    layer_1 = sigmoid(np.dot(layer_0,synapse_0))
    print("Sigmoid: {}",layer_1)

    # how much did we miss?
    # (soll) - (ist)
    layer_1_error = layer_1 - y

    print("Derivate or Back Propagation: ",sigmoid_output_to_derivative(layer_1))
    print("layer_1_error: ",layer_1_error)
    # multiply how much we missed by the 
    # slope of the sigmoid at the values in l1
    layer_1_delta = layer_1_error * sigmoid_output_to_derivative(layer_1)
    print("layer_1_delta:",layer_1_delta)
    #print('layer_1_delta: {}',layer_1_delta)
    synapse_0_derivative = np.dot(layer_0.T,layer_1_delta)
    print('synapse_0_derivative:',synapse_0_derivative)

    # update weights
    synapse_0 -= 10 * synapse_0_derivative
    print('synapse_0_ {}',synapse_0)

print("______________ END __________________") 
print("Output After Training: ")
print(layer_1)
```
